# Creacion de tablas

```python


create table Clientes (
        ClienteId int not null,
        nombre_cl varchar(20),
        ApCli varchar(20),
        AmCli varchar(20),
        fecha_nacimientoC datetime,
        correo nvarchar(30),
        Direccion_cl varchar(50),
        telefonoCli char(10)
)
GO


create table Empleados(
    RFC char(13) not null,
    Nombre_Emp varchar(20),
    ApEm varchar(20),
    AmEm varchar(20),
    fecha_nacimientoE datetime,
    fecha_Contratacion datetime,
    puesto varchar(20),
    Sueldo numeric(10,3),
    telefonoEmp char(10)

)

create table Pedidos(
    Pedidoid int not null,
    Fecha_Pedido datetime, 
    ClienteId int not null,
    RFC char(13) not null
)
go



create table Categorias(
    CategoriaId int not null,
    Nombre_c varchar(20),
    Descripcion_c varchar(50)
)
go


create  table Proveedores(
    ProveedorId int not null,
    nombre_Pr varchar(20),
    ApPr varchar(20),
    AmPr varchar(20),
    Direccion_pr varchar(50)
) 

create table Productos(
    ProductoId int not null,
    nombre_p varchar(20),
    Descripcion_p varchar(50),
    stock int,
    Precio numeric(10,2),
    ProveedorId int not null
)
go


create table DetallePedido(
    pedidoId int not null,
    ProductoId int not null,
    Descuento numeric(3,2),
    Cantidad int,
    PrecioUnitario numeric(10,2),
    ProductoId int not null,
    PedidoId int not null
)
GO



```

# CONSTRAINT Primary Key

```python

Alter table Clientes
add CONSTRAINT pk_Cliente PRIMARY KEY(ClienteId)
GO

Alter table Empleados 
add constraint pk_Emp PRIMARY KEY(RFC)
GO

Alter table Pedidos 
add CONSTRAINT pk_Pedido Primary key(PedidoId)
GO

alter table proveedores 
add constraint pk_prov PRIMARY KEY(Proveedorid)
go

alter table Categorias 
add CONSTRAINT pk_cat PRIMARY KEY(categoriaId)
GO

Alter table Productos 
add CONSTRAINT pk_Produc PRIMARY KEY(Productid)
GO

Alter table DetallePedido
add constraint pk_Detalle primary key(pedidoId, ProductoId)
```

# CONSTRAINT FOREIGN KEY

```python
Alter table Pedidos
add Constraint fk_RFC FOREIGN KEY references Empleados(RFC)
go

Alter table Produtos
add constraint fk_Prov FOREIGN KEY references Proveedores(Provedorid),
    CONSTRAINT fk_Cat FOREIGN KEY REFERENCES Categorias(CategoriaId)
GO

Alter table DetallePedido
add CONSTRAINT fk_Produc FOREIGN KEY REFERENCES Productos(productoId),
    constraint fk_Pedido foreign key REFERENCES Pedidos(pedidoId)
    GO
```

# Constraint Check, default, unique

```python
Alter table Clientes
add constraint Ck_telefonoCl CHECK (LEN(telefonoCli)=10),
    constraint Ck_MayorC CHECK(datediff(yyyy, fecha_nacimientoC, getdate())>=18),
    CONSTRAINT DF_DIRECCION DEFAULT 'Sin domicilio' for direccion_cl,
    constraint un_TelefonoEm unique(telefonoCli)
go


Alter table Empleados 
    add constraint Ck_telefonoEm CHECK (LEN(telefonoEmp)=10),
        constraint Ck_RFCEm CHECK (LEN(RFC)=13),
        constraint Ck_MayorE CHECK(datediff(yyyy, fecha_nacimientoE, Fecha_contratacion)>=18),
        constraint un_TelefonoEm unique(telefonoEmp),
        constraint df_puesto DEFAULT 'Vendedor' for puesto
GO

alter table Proveedores
add constraint Ck_telefonopr CHECK (LEN(telefonopr)=10),
    constraint un_Telefonopr unique(telefono_pr)
GO


Alter table DetallePedido
add constraint Ck_PrecioU Check(PrecioUnitario>0),
    constraint Ck_Cantidad Check(Cantidad>0)

Alter table Productos
add CONSTRAINT Ck_Precio Check(Precio>0)

```
